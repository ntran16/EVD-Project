```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import (preprocessing, metrics, cross_validation)
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import scipy.io as sio

```

```python
# def get_score(datapath, datapath_y, score_type):
#     start = time.time()
for count in range(1,25):
    datapath = 'D:/Dropbox (Nate)/Rotation 3 Project/Data/FeatData_Agg/feature_%0.2dh.csv'%count
    datapath_y = 'D:/Dropbox (Nate)/Rotation 3 Project/Data/FeatData_Agg/target_%0.2dh.csv'%count
    X = pd.read_csv(datapath)
    X = np.nan_to_num(X) # replace nan with 0
    scaler = preprocessing.MinMaxScaler().fit(X)
    X = scaler.transform(X)

    Y = pd.read_csv(datapath_y)
    Y = np.ravel(Y)

    #     loo = LeaveOneOut()
    #     fold = StratifiedKFold(Y, n_folds=5, shuffle=True, random_state=40)
    # #     SP = StratifiedShuffleSplit(n_splits=20, test_size=0.2, random_state=0)
    #     ## Dynamic Shuffle
    #     SP = StratifiedShuffleSplit(n_splits=20, test_size=.2, random_state=40)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0, stratify = Y)
    lda = LinearDiscriminantAnalysis(solver='eigen', shrinkage=.5)
    lda.fit(X_train, Y_train)
    X_train_new = lda.transform(X_train)
    X_test_new = lda.transform(X_test)
    
    plt.figure()
    colors = ['red', 'green']

    bins = np.linspace(min(X_train_new), max(X_train_new), 20)

    # fig = plt.figure(figsize=(10,4))
    # plt.subplot(1, 2, 1)
    # plt.title("Feature 1 - Pass")
    # plt.xlabel("Pass")
    # plt.hist(X_r2[Y == 0], bins = bins, color = 'red')
    # plt.subplot(1, 2, 2)
    # plt.title("Feature 1 - Fail")
    # plt.xlabel("Fail")
    # plt.hist(X_r2[Y == 1],bins = bins, color = 'green')
    # plt.show()

    plt.hist(X_train_new[Y_train==0], bins, alpha=0.5, label='Pass')
    plt.hist(X_train_new[Y_train==1], bins, alpha=0.5, label='Fail')
    plt.legend(loc='upper right')
    plt.suptitle('%s'%count)
    plt.show()

#     k = LinearDiscriminantAnalysis()
#     k.fit(X_train_new, Y_train)
#     guess = k.decision_function(X_test_new)
    
#     fpr, tpr, thresholds = roc_curve(Y_test, guess)
# #     fpr, tpr, thresholds = roc_curve(Y_true[i], Y_proba[i][:,1])
#     roc_auc = auc(fpr, tpr)

#     print('the score of %s is %f'%(count,roc_auc))
```
